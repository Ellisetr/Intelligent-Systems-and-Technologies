```python
import pandas as pd
import numpy as np
from itertools import *
```

```python
small_df = pd.read_excel('https://docs.google.com/spreadsheets/d/1aZTDk8rHtzlCi09DDvI-w1wEERD2aUoL/export?format=xlsx',header=None)
small_df
```

Подготовка данных

```python
trans_from_df = small_df.values.tolist()
transactions = []
for rows in trans_from_df:
  clean_row = []
  for items in rows:
    if items is not np.nan:
      clean_row.append(items)
  transactions.append(clean_row) 

minsup = 0.2
minconf = 0.6
real_minsup = len(transactions)*minsup
```

Пускай у нас будет minsup = 0.2, minconf = 0.8

```python
unique = {}
unique['Single Items'] = {}
for lines in transactions:
  for items in lines:
    if items not in unique['Single Items']:
      unique['Single Items'][items] = 1
    else:
      unique['Single Items'][items] += 1
print('Общее количество элементов в таблице')
for keys in unique['Single Items']:
  print(keys,':',unique['Single Items'][keys])
```

```python
unique['Double Items']={}
for lines in transactions:
  for items in list(combinations((sorted(lines)),2)):
    if items not in unique['Double Items']:
      unique['Double Items'][items] = 1
    else:
      unique['Double Items'][items] += 1

to_del = []
for keys in unique['Double Items']:
  if unique['Double Items'][keys] < real_minsup:
    to_del.append(keys)
for keys in to_del:
  unique['Double Items'].pop(keys)

print('Общее количество 2-предметных наборов в таблице, которые подходят под MinSup')
for keys in sorted(unique['Double Items']):
  print(keys,':',unique['Double Items'][keys])
```

```python
unique['Triple Items'] = {}

for lines in transactions:
  for items in list(combinations((sorted(lines)),3)):
    if items not in unique['Triple Items']:
      unique['Triple Items'][items] = 1
    else:
      unique['Triple Items'][items] += 1

to_del = []
for keys in unique['Triple Items']:
  if unique['Triple Items'][keys] < real_minsup:
    to_del.append(keys)
for keys in to_del:
  unique['Triple Items'].pop(keys)

print('Общее количество 3-предметных наборов в таблице, которые подходят под MinSup')
for keys in sorted(unique['Triple Items']):
  print(keys,':',unique['Triple Items'][keys])
```

```python
unique['Quad Items'] = {}

for lines in transactions:
  for items in list(combinations((sorted(lines)),4)):
    if items not in unique['Quad Items']:
      unique['Quad Items'][items] = 1
    else:
      unique['Quad Items'][items] += 1

to_del = []
for keys in unique['Quad Items']:
  if unique['Quad Items'][keys] < real_minsup:
    to_del.append(keys)
for keys in to_del:
  unique['Quad Items'].pop(keys)

print('Общее количество 4-предметных наборов в таблице, которые подходят под MinSup')
for keys in sorted(unique['Quad Items']):
  print(keys,':',unique['Quad Items'][keys])
```

```python
# Для 2-ух элементов
rules = []
for keys in unique['Double Items']:
  conf = round(unique['Double Items'][keys]/unique['Single Items'][keys[1]],4)
  sup = round(unique['Double Items'][keys]/len(transactions),4)
  if conf >= minconf:
    rules.append([{keys[1]},{keys[0]},conf,sup])
    
  conf = round(unique['Double Items'][keys]/unique['Single Items'][keys[0]],4)
  sup = round(unique['Double Items'][keys]/len(transactions),4)
  if conf >= minconf:
    rules.append([{keys[0]},{keys[1]},conf,sup])
```

```python
rules1 = []
for keys in unique['Triple Items']:
  for lines in combinations(keys,2):
    conf = round(unique['Triple Items'][keys]/unique['Double Items'][lines],4)
    sup = round(unique['Triple Items'][keys]/len(transactions),4)
    if conf >= minconf:
      rules1.append([set(lines),(set(keys)-set(lines)),conf,sup])

  for lines in combinations(keys,2):
    conf = round(unique['Triple Items'][keys]/unique['Single Items'][list(set(keys)-set(lines))[0]],4)
    sup = round(unique['Triple Items'][keys]/len(transactions),4)
    if conf >= minconf:
      rules1.append([(set(keys)-set(lines)),set(lines),conf,sup])
```

```python
def print_rules(rules):
  for line in rules:
    print("{} -> {} (conf: {}, supp: {})".format(line[0],line[1],line[2],line[3]))

print_rules(rules)
print_rules(rules1)
```
